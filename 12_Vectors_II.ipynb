```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
a = np.array([2, 1])
b = np.array([-1, 1])
sum_ab = a + b

# Create a plot
plt.figure()
ax = plt.gca()

# Plot vector a
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector a')

# Plot vector b
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector b')

# Plot the sum of vectors a and b
ax.quiver(0, 0, sum_ab[0], sum_ab[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a + Vector b')

# Set the limits of the plot
plt.xlim(-2, 3)
plt.ylim(-1, 3)

# Add grid, labels, and legend
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)

# Add title
plt.title('Vectors a, b, and Their Sum')

# Show the plot
plt.show()
```

```python
import numpy as np

def are_perpendicular(vector1, vector2):
    """
    Checks if two vectors are perpendicular.
    """
    # Calculate the dot product
    dot_product = np.dot(vector1, vector2)

    # If the dot product is zero, the vectors are perpendicular
    if dot_product == 0:
        print("The vectors are perpendicular.")
    else:
        print("The vectors are not perpendicular.")

# Example usage
vector1 = np.array([2, 1])
vector2 = np.array([-1, 1])

are_perpendicular(vector1, vector2)
```

```python
#5
import numpy as np

# Define the vectors
a = np.array([4, 2, 1])
b = np.array([1, 3, 2])

# Calculate the dot product
dot_product = np.dot(a, b)

# Calculate the magnitudes of the vectors
magnitude_a = np.linalg.norm(a)
magnitude_b = np.linalg.norm(b)

# Calculate the cosine of the angle
cos_alpha = dot_product / (magnitude_a * magnitude_b)

# Calculate the angle in radians
angle_radians = np.arccos(cos_alpha)

# Convert the angle to degrees
angle_degrees = np.degrees(angle_radians)

# Print the result
print(f"The angle between the vectors is: {angle_degrees:.2f}°")
```
