#1.1
The given differential equation is:


y'(x) = y


This is a separable differential equation. To solve it, follow these steps:

### Step 1: Rearrange the equation
We can separate the variables \(y\) and \(x\) by dividing both sides by \(y\) (assuming $y \neq 0\$ :

$
\frac{y'(x)}{y} = 1
$

### Step 2: Integrate both sides
Next, integrate both sides with respect to \(x\):

$
\int \frac{1}{y} \, dy = \int 1 \, dx
$

The integral of $ \frac{1}{y} \, dy $ is $ \ln|y| $, and the integral of 1 with respect to \(x\) is \(x\). Therefore:

$
\ln|y| = x + C
$

where \(C\) is the constant of integration.

### Step 3: Solve for \(y\)
To isolate \(y\), exponentiate both sides:

$
|y| = e^{x + C}
$

This can be written as:

$
|y| = e^C e^x
$

Let $e^C $ be a new constant, say $C_1$, so that:


$ |y| = C_1 e^x $


Thus, the general solution for \(y\) is:

$
y = C_1 e^x
$

where $C_1$ can be any constant, including negative values.

### Final solution:

$
y(x) = C_1 e^x
$

This is the general solution to the differential equation $y'(x) = y$.

The given differential equation is:

$
y'(x) = \frac{1}{2} y(x)
$

This is a separable differential equation. Let's solve it step by step:

### Step 1: Rearrange the equation
We can separate the variables \(y\) and \(x\) by multiplying both sides by \(2 \, dx\) and dividing both sides by \(y(x)\):

$
\frac{2 \, dy}{y} = dx
$

### Step 2: Integrate both sides
Now, integrate both sides:

$
\int \frac{2}{y} \, dy = \int 1 \, dx
$

The integral of $ \frac{2}{y} \, dy $ is $ 2 \ln|y| $, and the integral of 1 with respect to \(x\) is \(x\). Therefore:

$
2 \ln|y| = x + C
$

where \(C\) is the constant of integration.

### Step 3: Solve for \(y\)
To isolate \(y\), divide both sides by 2:

$
\ln|y| = \frac{x}{2} + \frac{C}{2}
$

Exponentiate both sides:

$
|y| = e^{\frac{x}{2} + \frac{C}{2}}
$

We can rewrite $e^{\frac{C}{2}}$ as a new constant, say $C_1$:
$
|y| = C_1 e^{\frac{x}{2}}
$

Thus, the general solution for y is:
$
y = \pm C_1 e^{\frac{x}{2}}
$

where $C_1$ is a constant, and we include the $pm $ to account for both positive and negative values of \(y\).

### Final solution:

$
y(x) = C_1 e^{\frac{x}{2}}
$

where $C_1$ is a constant.

To solve the first-order differential equation:

$
\frac{dy}{dx} = \frac{x}{y}
$

using the method of separation of variables, follow these steps:

### Step 1: Rearrange the equation
Separate the variables \(x\) and \(y\) so that all terms involving \(y\) are on one side and all terms involving \(x\) are on the other side. Multiply both sides of the equation by \(y \, dx\):

$
y \, dy = x \, dx
$

### Step 2: Integrate both sides
Now, integrate both sides:

$
\int y \, dy = \int x \, dx
$

The integral of $ \int y \, dy$ is $ \frac{y^2}{2} $, and the integral of $ \int x \, dx$ is $\frac{x^2}{2}$. Therefore:

$
\frac{y^2}{2} = \frac{x^2}{2} + C
$

where $C$ is the constant of integration.

### Step 3: Solve for y
Multiply through by 2 to simplify:

$
y^2 = x^2 + 2C
$

Now, take the square root of both sides:

$
y = \pm \sqrt{x^2 + 2C}
$

### Final solution:
Thus, the general solution to the differential equation is:

$
y(x) = \pm \sqrt{x^2 + 2C}
$

where \(C\) is an arbitrary constant.

To solve the differential equation:

$
\frac{dy}{dx} = \frac{y}{x}
$

using the method of separation of variables, follow these steps:

### Step 1: Rearrange the equation
Separate the variables \(y\) and \(x\) so that all terms involving \(y\) are on one side and all terms involving \(x\) are on the other side. Multiply both sides by $ \frac{dx}{y} $:

$
\frac{dy}{y} = \frac{dx}{x}
$

### Step 2: Integrate both sides
Now, integrate both sides:

$
\int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx
$

The integral of $ \frac{1}{y} \, dy $ is $ \ln|y| $, and the integral of $ \frac{1}{x} \, dx $ is $ \ln|x| $. Therefore:

$
\ln|y| = \ln|x| + C
$

where $C$ is the constant of integration.

### Step 3: Solve for $y$
Exponentiate both sides to eliminate the logarithms:

$
|y| = e^{\ln|x| + C}
$

This can be rewritten as:

$
|y| = e^{\ln|x|} \cdot e^C
$

Since $e^{\ln|x|} = |x|$, we have:

$
|y| = C_1 |x|
$

where $C_1 = e^C$ is a new constant. Therefore, the general solution for $y$ is:

$
y = \pm C_1 x
$

### Final solution:

$
y(x) = C_1 x
$

where $C_1$ is an arbitrary constant.

# 3.1


We are given the second-order ordinary differential equation:

$
y''(x) + y'(x) = 0
$

with boundary conditions:

$
y(0) = 2 \quad \text{and} \quad y'(0) = -1
$

### Step 1: Solve the homogeneous equation
This is a linear homogeneous second-order differential equation. To solve it, we first write it as:

$
y''(x) + y'(x) = 0
$

We assume a solution of the form $ y(x) = e^{rx} $, where $r$ is a constant. Substituting this into the differential equation:

$
r^2 e^{rx} + r e^{rx} = 0
$

Factor out $ e^{rx} $:

$
e^{rx} (r^2 + r) = 0
$

Since $ e^{rx} \neq 0 $, we have:

$
r^2 + r = 0
$

This factors as:

$
r(r + 1) = 0
$

Thus, the two possible values for $r$ are:

$
r = 0 \quad \text{or} \quad r = -1
$

### Step 2: General solution
The general solution to the differential equation is:

$
y(x) = C_1 e^{0x} + C_2 e^{-x} = C_1 + C_2 e^{-x}
$

where $C_1$ and $C_2$ are constants to be determined from the boundary conditions.

### Step 3: Apply the boundary conditions

1. **Use $ y(0) = 2 $:**

Substitute $x = 0$ into the general solution:

$
y(0) = C_1 + C_2 e^{0} = C_1 + C_2
$

Since $y(0) = 2$, we have:

$
C_1 + C_2 = 2 \quad \text{(Equation 1)}
$

2. **Use $ y'(0) = -1 $:**

First, compute $ y'(x) $:

$
y'(x) = -C_2 e^{-x}
$

Substitute $x = 0$:

$
y'(0) = -C_2 e^{0} = -C_2
$

Since $y'(0) = -1$, we have:

$
-C_2 = -1 \quad \Rightarrow \quad C_2 = 1 \quad \text{(Equation 2)}
$

### Step 4: Solve for $C_1$

From Equation 1:

$
C_1 + C_2 = 2
$

Substitute $C_2 = 1$:

$
C_1 + 1 = 2 \quad \Rightarrow \quad C_1 = 1
$

### Step 5: Final solution

Substitute $C_1 = 1$ and $C_2 = 1$ into the general solution:

$
y(x) = 1 + e^{-x}
$

### Final Solution:

$
y(x) = 1 + e^{-x}
$

This is the solution to the differential equation $ y''(x) + y'(x) = 0 $ with the given boundary conditions $ y(0) = 2 $  and $ y'(0) = -1 $

We are given the second-order differential equation:

$
\frac{d^2 y(x)}{dx^2} - \omega^2 y(x) = 0
$

### Step 1: Solve the homogeneous equation
This is a linear homogeneous second-order differential equation. To solve it, assume a solution of the form:

$
y(x) = e^{rx}
$

Substitute $y(x) = e^{rx}$ into the differential equation:

$
r^2 e^{rx} - \omega^2 e^{rx} = 0
$

Factor out $e^{rx}$:

$
e^{rx} (r^2 - \omega^2) = 0
$

Since $e^{rx} \neq 0$, we have:

$
r^2 - \omega^2 = 0
$

Solve for $r$:

$
r = \pm \omega
$

### Step 2: General solution
The general solution to the differential equation is:

$
y(x) = C_1 e^{\omega x} + C_2 e^{-\omega x}
$

where $C_1$ and $C_2$ are arbitrary constants.
