## Solving the System of Linear Equations using Gaussian Elimination

We are given the system of equations:

$\begin{cases}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{cases}$

### **Step 1: Write the Augmented Matrix**
The system can be represented as the augmented matrix:

$
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
2 & 1 & 1 & 0 \\
3 & 2 & 1 & 1
\end{array}
\right]
$

### **Step 2: Convert to Upper Triangular Form**

#### **Eliminate the first column below the pivot**
- Subtract **2 × (Row 1)** from **Row 2**
- Subtract **3 × (Row 1)** from **Row 3**

New matrix:

$
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & -4 & 7 & -11
\end{array}
\right]
$

#### **Eliminate the second column below the pivot**
- Multiply **Row 2** by \(-\frac{4}{3}\) and add to **Row 3**

New matrix:

$
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & 0 & \frac{4}{3} & \frac{-4}{3}
\end{array}
\right]
$

Multiply **Row 3** by \( \frac{3}{4} \) to make the pivot 1:

$
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & 0 & 1 & -1
\end{array}
\right]
$

### **Step 3: Back Substitution**
From **Row 3**:
$
z = -1
$

From **Row 2**:
$
-3y + 5(-1) = -8
$
$
-3y - 5 = -8
$
$
-3y = -3
$
$
y = 1
$

From **Row 1**:
$
x + 2(1) - 2(-1) = 4
$
$
x + 2 + 2 = 4
$
$
x = 0
$

### **Final Answer**
$
x = 0, \quad y = 1, \quad z = -1
$

\begin{cases}
x + y + z - t = 2 \quad \text{(1)} \\
2x + y + z = 3 \quad \text{(2)} \\
-x + z - t = 0 \quad \text{(3)} \\
3x + 2y - z + 2t = -1 \quad \text{(4)}
\end{cases}
$

---

### Step 1: Write the augmented matrix
The augmented matrix for the system is:

$
\begin{bmatrix}
1 & 1 & 1 & -1 & | & 2 \\
2 & 1 & 1 & 0 & | & 3 \\
-1 & 0 & 1 & -1 & | & 0 \\
3 & 2 & -1 & 2 & | & -1
\end{bmatrix}
$

---

### Step 2: Perform row operations to get the matrix into row-echelon form

#### Step 2.1: Make the first element of the first row a pivot (already 1).

#### Step 2.2: Eliminate the first element of the second, third, and fourth rows.

- **Row 2**: Subtract $ 2 \times \text{Row 1} $ from Row 2:
  $
  \text{Row 2} = \text{Row 2} - 2 \times \text{Row 1}
  $
  Result:
  $
  \begin{bmatrix}
  0 & -1 & -1 & 2 & | & -1
  \end{bmatrix}
  $

- **Row 3**: Add Row 1 to Row 3:
  $
  \text{Row 3} = \text{Row 3} + \text{Row 1}
  $
  Result:
  $
  \begin{bmatrix}
  0 & 1 & 2 & -2 & | & 2
  \end{bmatrix}
  $

- **Row 4**: Subtract $ 3 \times \text{Row 1} $ from Row 4:
  $
  \text{Row 4} = \text{Row 4} - 3 \times \text{Row 1}
  $
  Result:
  $
  \begin{bmatrix}
  0 & -1 & -4 & 5 & | & -7
  \end{bmatrix}
  $

The updated matrix is:

$
\begin{bmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & -1 & -1 & 2 & | & -1 \\
0 & 1 & 2 & -2 & | & 2 \\
0 & -1 & -4 & 5 & | & -7
\end{bmatrix}
$

---

#### Step 2.3: Make the second element of the second row a pivot.

- Divide Row 2 by $-1$
  \text{Row 2} = -\text{Row 2}
  $
  Result:
  $
  \begin{bmatrix}
  0 & 1 & 1 & -2 & | & 1
  \end{bmatrix}
  $

The updated matrix is:

$
\begin{bmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & 1 & 1 & -2 & | & 1 \\
0 & 1 & 2 & -2 & | & 2 \\
0 & -1 & -4 & 5 & | & -7
\end{bmatrix}
$

---

#### Step 2.4: Eliminate the second element of the third and fourth rows.

- **Row 3**: Subtract Row 2 from Row 3:
  $
  \text{Row 3} = \text{Row 3} - \text{Row 2}
  $
  Result:
  $
  \begin{bmatrix}
  0 & 0 & 1 & 0 & | & 1
  \end{bmatrix}
  $

- **Row 4**: Add Row 2 to Row 4:
  $
  \text{Row 4} = \text{Row 4} + \text{Row 2}
  $
  Result:
  $
  \begin{bmatrix}
  0 & 0 & -3 & 3 & | & -6
  \end{bmatrix}
  $

The updated matrix is:

$
\begin{bmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & 1 & 1 & -2 & | & 1 \\
0 & 0 & 1 & 0 & | & 1 \\
0 & 0 & -3 & 3 & | & -6
\end{bmatrix}
$

---

#### Step 2.5: Make the third element of the third row a pivot (already 1).

#### Step 2.6: Eliminate the third element of the fourth row.

- **Row 4**: Add \( 3 \times \text{Row 3} \) to Row 4:
  $
  \text{Row 4} = \text{Row 4} + 3 \times \text{Row 3}
  $
  Result:
  $
  \begin{bmatrix}
  0 & 0 & 0 & 3 & | & -3
  \end{bmatrix}
  $

The updated matrix is:

$
\begin{bmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & 1 & 1 & -2 & | & 1 \\
0 & 0 & 1 & 0 & | & 1 \\
0 & 0 & 0 & 3 & | & -3
\end{bmatrix}
$

---

### Step 3: Back-substitution

#### Step 3.1: Solve for \( t \):
From Row 4:
$
3t = -3 \implies t = -1
$

#### Step 3.2: Solve for \( z \):
From Row 3:
$
z = 1
$

#### Step 3.3: Solve for \( y \):
From Row 2:
$
y + z - 2t = 1 \implies y + 1 - 2(-1) = 1 \implies y + 3 = 1 \implies y = -2
$

#### Step 3.4: Solve for \( x \):
From Row 1:
$
x + y + z - t = 2 \implies x - 2 + 1 - (-1) = 2 \implies x = 2
$

---

### Final Solution:
$
x = 2, \quad y = -2, \quad z = 1, \quad t = -1
$
