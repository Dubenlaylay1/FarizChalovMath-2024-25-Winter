```python
#1
import numpy as np
import matplotlib.pyplot as plt

# Define f(x)
def f(x):
    return x**2

# Evaluate f(2)
x_val = 2
f_val = f(x_val)
print(f"f(2) = {f_val}")

# Plot f(x)
x = np.linspace(0, 10, 500)
plt.figure(figsize=(6, 4))
plt.plot(x, f(x), label="f(x) = x^2", color="blue")
plt.scatter([x_val], [f_val], color="blue", label=f"f(2) = {f_val:.2f}")
plt.title("Plot of f(x) = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# Define g(x)
def g(x):
    return np.sqrt(x)

# Evaluate g(2)
x_val = 2
g_val = g(x_val)
print(f"g(2) = {g_val}")

# Plot g(x)
x = np.linspace(0, 10, 500)
plt.figure(figsize=(6, 4))
plt.plot(x, g(x), label="g(x) = sqrt(x)", color="orange")
plt.scatter([x_val], [g_val], color="orange", label=f"g(2) = {g_val:.2f}")
plt.title("Plot of g(x) = sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt

# Define h(x)
def h(x):
    return 1/x

# Evaluate h(2)
x_val = 2
h_val = h(x_val)
print(f"h(2) = {h_val}")

# Plot h(x)
x = np.linspace(0.1, 10, 500)  # Avoid x = 0 to prevent division by zero
plt.figure(figsize=(6, 4))
plt.plot(x, h(x), label="h(x) = 1/x", color="green")
plt.scatter([x_val], [h_val], color="green", label=f"h(2) = {h_val:.2f}")
plt.title("Plot of h(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
#4import numpy as np
import matplotlib.pyplot as plt

# Define j(x)
def j(x):
    return np.sin(x)

# Evaluate j(2)
x_val = 2
j_val = j(x_val)
print(f"j(2) = {j_val}")

# Plot j(x)
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # Wider range for sin(x)
plt.figure(figsize=(6, 4))
plt.plot(x, j(x), label="j(x) = sin(x)", color="red")
plt.scatter([x_val], [j_val], color="red", label=f"j(2) = {j_val:.2f}")
plt.title("Plot of j(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```
