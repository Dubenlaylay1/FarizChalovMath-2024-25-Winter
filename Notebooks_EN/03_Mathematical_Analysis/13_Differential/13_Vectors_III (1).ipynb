```python
#1
import matplotlib.pyplot as plt

# Coordinates of points A and B
A = (-1, 2)
B = (3, -2)

# Ratio m:n
m, n = 1, 3

# Calculate coordinates of point P that divides the segment in the ratio 1:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(-1, 2)')
plt.scatter(*B, color='red', label='B(3, -2)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.3, A[1]+0.2, 'A(-1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1]-0.3, 'B(3, -2)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1]-0.3, f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 1:3')
plt.legend()

# Displaying the graph
plt.show()
```

To divide the line segment connecting points $ A(-1, 2) $ and $ B(3, -2) $ in the ratio $ 1:3 $, we use the **section formula**:

### Section Formula
If a point $ P(x, y) $ divides the segment joining $ A(x_1, y_1) $ and $ B(x_2, y_2) $ in the ratio $ m:n $, its coordinates are:
$
x = \frac{m x_2 + n x_1}{m + n}, \quad y = \frac{m y_2 + n y_1}{m + n}
$

### Given:
- $ A(-1, 2) $
- $ B(3, -2) $
- Ratio $ m:n = 1:3 $

### Calculation:
1. **Identify values**:
   - $ x_1 = -1 $, $ y_1 = 2 $
   - $ x_2 = 3 $, $ y_2 = -2 $
   - $ m = 1 $, $ n = 3 $

2. **Compute $ x $**:
   $
   x = \frac{(1 \cdot 3) + (3 \cdot -1)}{1 + 3} = \frac{3 - 3}{4} = 0
   $

3. **Compute $ y $**:
   $
   y = \frac{(1 \cdot -2) + (3 \cdot 2)}{1 + 3} = \frac{-2 + 6}{4} = 1
   $

### Final Answer:
The point dividing $ AB $ in the ratio $ 1:3 $ is:
$
\boxed{(0, 1)}
$

```python
#2
import matplotlib.pyplot as plt

# Vector a
a = (3, 4)

# Unit vectors for the x-axis and y-axis
i = (1, 0)  # Unit vector of x-axis
j = (0, 1)  # Unit vector of y-axis

# Dot product calculations
dot_a_i = a[0] * i[0] + a[1] * i[1]  # Projection onto x-axis
dot_a_j = a[0] * j[0] + a[1] * j[1]  # Projection onto y-axis

# Projections
proj_OX = (dot_a_i * i[0], dot_a_i * i[1])  # (3, 0)
proj_OY = (dot_a_j * j[0], dot_a_j * j[1])  # (0, 4)

# Plotting the graph
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (3, 4)')

# Plot projections
plt.quiver(0, 0, proj_OX[0], proj_OX[1], angles='xy', scale_units='xy', scale=1, color='red', label='Projection on OX (3, 0)')
plt.quiver(0, 0, proj_OY[0], proj_OY[1], angles='xy', scale_units='xy', scale=1, color='green', label='Projection on OY (0, 4)')

# Mark the points for vector tips and projections
plt.scatter([a[0], proj_OX[0], proj_OY[0]], [a[1], proj_OX[1], proj_OY[1]], color=['blue', 'red', 'green'])

# Annotations
plt.text(a[0] + 0.2, a[1], f'a({a[0]}, {a[1]})', fontsize=10, color='blue')
plt.text(proj_OX[0] + 0.2, proj_OX[1], f'OX({proj_OX[0]}, {proj_OX[1]})', fontsize=10, color='red')
plt.text(proj_OY[0] + 0.2, proj_OY[1], f'OY({proj_OY[0]}, {proj_OY[1]})', fontsize=10, color='green')

# Setting the axes and grid
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Vector Projection onto OX and OY')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Display the graph
plt.show()
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np

# Vectors a and b
a = np.array([2, 3])
b = np.array([1, 1])

# Magnitude of vector b
b_magnitude = np.sqrt(b[0]**2 + b[1]**2)

# Unit vector of b
b_unit = b / b_magnitude

# Dot product of a and b_unit
dot_a_b_unit = np.dot(a, b_unit)

# Projection of a onto b
proj_b_a = dot_a_b_unit * b_unit

# Plotting
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (2, 3)')

# Plot vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Vector b (1, 1)')

# Plot projection of a onto b
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Projection of a onto b ({proj_b_a[0]:.2f}, {proj_b_a[1]:.2f})')

# Annotations
plt.text(a[0] + 0.2, a[1], 'a (2, 3)', fontsize=10, color='blue')
plt.text(b[0] + 0.2, b[1], 'b (1, 1)', fontsize=10, color='orange')
plt.text(proj_b_a[0] + 0.2, proj_b_a[1], f'Projection ({proj_b_a[0]:.2f}, {proj_b_a[1]:.2f})', fontsize=10, color='green')

# Set limits and grid
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Projection of Vector a onto Vector b')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show plot
plt.show()
```
