```python
#1
import sympy as sp

# Define the variables
x, y = sp.symbols("x y")

# Center and radius of the circle
h, k = 1, 2  # center coordinates
r = 3  # radius

# Equation of the circle
circle_equation = (x - h)**2 + (y - k)**2 - r**2

# Display the equation
print("The equation of the circle is:")
sp.pretty_print(sp.Eq(circle_equation, 0))
```

## Finding the Equation of a Circle

The equation of a circle with center \( (h, k) \) and radius \( r \) is given by:

$
(x - h)^2 + (y - k)^2 = r^2
$

### Given:
- Center: $ A(1,2) $ → $ h = 1, k = 2 $
- Radius: $ r = 3 $

### Substituting the values:

$
(x - 1)^2 + (y - 2)^2 = 3^2
$

$
(x - 1)^2 + (y - 2)^2 = 9
$

### Final Answer:
$
(x - 1)^2 + (y - 2)^2 = 9
$

## Finding the Equation of a Parabola

The equation of a parabola that intersects the \( x \)-axis at points \( x = 2 \) and \( x = 4 \) can be written in factored form as:

$
y = a(x - 2)(x - 4)
$

where $ a $ is a constant to be determined.

### Step 1: Use Given Point $ (x, y) = (3,1) $
Since the parabola passes through $ (3,1) $, we substitute $ x = 3 $ and $ y = 1 $ into the equation:

$
1 = a(3 - 2)(3 - 4)
$

$
1 = a(1)(-1)
$

$
1 = -a
$

$
a = -1
$

### Step 2: Write the Final Equation
Substituting $ a = -1 $ back into the equation:

$
y = - (x - 2)(x - 4)
$

Expanding:

$
y = - (x^2 - 6x + 8)
$

$
y = -x^2 + 6x - 8
$

### Final Answer:
$
y = -x^2 + 6x - 8
$

## Finding the Center of the Ellipse

We are given the equation of the ellipse:

$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$

### Step 1: Rearranging the Terms
Group the $ x $-terms and $ y $-terms together:

$
(x^2 - 4x) + 4(y^2 - 4y) = -16
$

### Step 2: Completing the Square

#### Completing the square for $ x $-terms:
- Take half of $-4$, square it:  
  $
  \left(\frac{-4}{2}\right)^2 = 4
  $
- Add and subtract 4 inside the parentheses:
$
  (x^2 - 4x) = (x - 2)^2 - 4
  $

#### Completing the square for $ y $-terms:
- Factor out 4 first:  
  $
  4(y^2 - 4y)
  $
- Take half of \(-4\), square it:  
  $
  \left(\frac{-4}{2}\right)^2 = 4
  $
- Add and subtract 4 inside the parentheses:

  $
  4(y^2 - 4y) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16
  $

### Step 3: Rewrite the Equation
Substituting the completed squares:

$
(x - 2)^2 - 4 + 4(y - 2)^2 - 16 = -16
$

$
(x - 2)^2 + 4(y - 2)^2 - 20 = -16
$

$
(x - 2)^2 + 4(y - 2)^2 = 4
$

### Step 4: Identify the Center
From the standard form of the ellipse equation:

$
(x - h)^2 + k(y - k)^2 = C
$

we identify the center as:

$

### Final Answer:
**The center of the ellipse is \( (2,2) \).**

## Finding the Slope \( m > 0 \) of the Tangent Line

We need to find the slope \( m > 0 \) of the line:

$
y = mx - 5
$

that is tangent to the circle with the equation:

$
x^2 + y^2 = 1
$

### Step 1: Substitute \( y = mx - 5 \) into the Circle's Equation
Substituting \( y = mx - 5 \) into \( x^2 + y^2 = 1 \):

$
x^2 + (mx - 5)^2 = 1
$

Expanding:

$
x^2 + (m^2x^2 - 10mx + 25) = 1
$

$
x^2 + m^2x^2 - 10mx + 25 = 1
$

$
(1 + m^2)x^2 - 10mx + 24 = 0
$

### Step 2: Find Condition for Tangency
For the line to be tangent to the circle, the quadratic equation in \( x \):

$
(1 + m^2)x^2 - 10mx + 24 = 0
$

must have exactly **one solution**, which means its discriminant must be **zero**:

$
\Delta = b^2 - 4ac = 0
$

where:
- $ a = 1 + m^2 $,
- $ b = -10m $,
- $ c = 24 $.

Computing the discriminant:

$
  (-10m)^2 - 4(1 + m^2)(24) = 0
$

$
100m^2 - 96(1 + m^2) = 0
$

$
100m^2 - 96 - 96m^2 = 0
$

$
(100m^2 - 96m^2) = 96
$

$
4m^2 = 96
$

$
m^2 = 24
$

$
m = \sqrt{24} = 2\sqrt{6}
$

Since we need \( m > 0 \), we take the positive root:

$
m = 2\sqrt{6}
$

### Final Answer:
$
m = 2\sqrt{6}
$

## Finding the Distance Between the Branches of the Hyperbola

We are given the equation of the hyperbola:

$
x^2 - y^2 = 1
$

### Step 1: Identify the Values of \( a \) and \( b \)
The standard form of the equation for a hyperbola centered at the origin is:

$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$

For the equation $ x^2 - y^2 = 1 $, we can rewrite it as:

$
\frac{x^2}{1} - \frac{y^2}{1} = 1
$

Here, $ a^2 = 1 $ and $ b^2 = 1 $, so:

$
a = 1 \quad \text{and} \quad b = 1
$

### Step 2: Calculate the Distance Between the Branches
The distance between the branches of the hyperbola is given by:

$
\text{Distance between branches} = 2a
$

Since $ a = 1 $, the distance between the branches is:

$
  2a = 2 \times 1 = 2
$

### Final Answer:
The distance between the branches of the hyperbola $ x^2 - y^2 = 1 $ is **2** units.
