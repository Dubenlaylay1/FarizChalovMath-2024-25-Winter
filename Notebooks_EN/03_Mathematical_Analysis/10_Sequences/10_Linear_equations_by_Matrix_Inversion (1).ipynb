```python
# ## Solving Systems of Linear Equations Using the Inverse Matrix Method

import sympy as sp

# Coefficient matrix
A = sp.Matrix([
    [1, 2, 3],
    [2, 0, 3],
    [0, 0, 3]
])

# Constant terms vector
B =  sp.Matrix([
    [5],
    [4],
    [13]
])
# Inverse matrix
A_inv = A.inv()

# Solution of the system of equations
X = A_inv * B
X
```

### 1. Express the system in matrix form:
The given system of equations can be written in the matrix form $ AX = B $, where:

$ A $ is the coefficient matrix:
$
A = \begin{pmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{pmatrix}
$

$ X $ is the variable matrix:
$
X = \begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
$

$ B $ is the constant matrix:
$
B = \begin{pmatrix}
5 \\
4 \\
3
\end{pmatrix}
$

So, we have:
$
A X = B
$
$
\begin{pmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
=
\begin{pmatrix}
5 \\
4 \\
3
\end{pmatrix}
$

---

### 2. Find the inverse of matrix \( A \) (\( A^{-1} \)):

For a 3x3 matrix, finding the inverse can be a bit involved. Here's the process:

- **Calculate the determinant of \( A \)**:

$
\text{det}(A) = 1 \times (2 \times 3 - 3 \times 0) - 2 \times (0 \times 3 - 3 \times 0) + 3 \times (0 \times 0 - 2 \times 0) = 1 \times 6 = 6
$

- **Find the matrix of cofactors**:
The cofactor of an element \( a_{ij} \) is \( (-1)^{i+j} \) times the determinant of the submatrix obtained by deleting the \( i \)-th row and \( j \)-th column.

$
C = \begin{pmatrix}
6 & 0 & 0 \\
-6 & 3 & 0 \\
0 & -3 & 2
\end{pmatrix}
$

- **Transpose the matrix of cofactors (also called the adjugate matrix)**:
$
\text{adj}(A) = \begin{pmatrix}
6 & -6 & 0 \\
0 & 3 & -3 \\
0 & 0 & 2
\end{pmatrix}
$

- **Multiply the adjugate matrix by $ 1/\text{det}(A) $ to get $ A^{-1} $**:
$
A^{-1} = \frac{1}{6} \times \begin{pmatrix}
6 & -6 & 0 \\
0 & 3 & -3 \\
0 & 0 & 2
\end{pmatrix}
= \begin{pmatrix}
1 & -1 & 0 \\
0 & \frac{1}{2} & -\frac{1}{2} \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
$

---

### 3. Multiply $ A^{-1} $ by $ B $ to find $ X $:

$
X = A^{-1} B = \begin{pmatrix}
1 & -1 & 0 \\
0 & \frac{1}{2} & -\frac{1}{2} \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
\begin{pmatrix}
5 \\
4 \\
3
\end{pmatrix}
$

$
X = \begin{pmatrix}
(1 \times 5) + (-1 \times 4) + (0 \times 3) \\
(0 \times 5) + \left(\frac{1}{2} \times 4\right) + \left(-\frac{1}{2} \times 3\right) \\
(0 \times 5) + (0 \times 4) + \left(\frac{1}{3} \times 3\right)
\end{pmatrix}
$

$
X = \begin{pmatrix}
5 - 4 \\
2 - \frac{3}{2} \\
1
\end{pmatrix}
= \begin{pmatrix}
1 \\
\frac{1}{2} \\
1
\end{pmatrix}
$

---

### 4. Solution:
Therefore, the solution to the system of equations is:

$
x = 1, \quad y = \frac{1}{2}, \quad z = 1
$

Since \(\text{det}(A) = 0\), \( A \) is **singular** and **no inverse exists**.
