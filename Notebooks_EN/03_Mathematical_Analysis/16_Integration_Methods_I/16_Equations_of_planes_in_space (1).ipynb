```python
#1
import numpy as np

# Define the points A, B, and C as numpy arrays
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Calculate two vectors in the plane
AB = B - A
AC = C - A

# Calculate the normal vector to the plane using the cross product
normal_vector = np.cross(AB, AC)

# Extract the coefficients of the plane equation (ax + by + cz + d = 0)
a, b, c = normal_vector

# Find d using point A
d = -np.dot(normal_vector, A)

# Print the equation of the plane
print(f"The equation of the plane is: {a}x + {b}y + {c}z + {d} = 0")
```

To find the equation of the plane passing through points \( A(1, 2, 3) \), \( B(3, 4, 5) \), and \( C(2, 1, 4) \), follow these steps:

---

### 1. **Find Two Vectors in the Plane**
   - **Vector \( \overrightarrow{AB} \):**  
     $
     \overrightarrow{AB} = B - A = (3-1, 4-2, 5-3) = (2, 2, 2)
     $
   - **Vector \( \overrightarrow{AC} \):**  
     $
     \overrightarrow{AC} = C - A = (2-1, 1-2, 4-3) = (1, -1, 1)
     $

---

### 2. **Compute the Cross Product (Normal Vector)**
   - The cross product $ \overrightarrow{AB} \times \overrightarrow{AC} $ gives a normal vector $ \mathbf{n} $ to the plane:  
     
     $\mathbf{n} = \begin{vmatrix}
     \mathbf{i} & \mathbf{j} & \mathbf{k} \\
     2 & 2 & 2 \\
     1 & -1 & 1
     \end{vmatrix}  
     
     $
     \mathbf{n} = \mathbf{i}(2 \cdot 1 - 2 \cdot (-1)) - \mathbf{j}(2 \cdot 1 - 2 \cdot 1) + \mathbf{k}(2 \cdot (-1) - 2 \cdot 1)
     $
     $
     \mathbf{n} = (4)\mathbf{i} - (0)\mathbf{j} + (-4)\mathbf{k} = (4, 0, -4)
     $
   - Simplify by dividing by 4:  
     $
     \mathbf{n} = (1, 0, -1)
     $

---

### 3. **Write the Plane Equation**
   - Use the normal vector $ \mathbf{n} = (1, 0, -1) $ and point $ A(1, 2, 3) $:  
     $
     1(x - 1) + 0(y - 2) - 1(z - 3) = 0
     $
   - Simplify:  
     $
     x - 1 - z + 3 = 0 \implies x - z + 2 = 0
     $

---

### 4. **Verification**
   - **Point \( A(1, 2, 3) \):**  
     $
     1 - 3 + 2 = 0 \quad \checkmark
     $
   - **Point \( B(3, 4, 5) \):**  
     $
     3 - 5 + 2 = 0 \quad \checkmark
     $
   - **Point $ C(2, 1, 4) $:**  
     $
     2 - 4 + 2 = 0 \quad \checkmark
     $

---

**Final Answer:**  
$
\boxed{x - z + 2 = 0}
$

```python
#2
import sympy as sp

# Define variables for the plane equation
x, y, z = sp.symbols('x y z')

# Normal vector coefficients
a, b, c = 2, 3, 4

# Point through which the plane passes
point = (1, 2, 3)

# Calculate 'd' using the point (substitute into the plane equation)
d = a * point[0] + b * point[1] + c * point[2]

# Define the equation of the plane
plane_equation = sp.Eq(a * x + b * y + c * z, d)

# Display the result
print("The equation of the plane is:", plane_equation)
```

To find the line of intersection of the two planes:

$
2x + 3y + 4z = 5 \quad \text{(Plane 1)}
$
$
3x + 4y + 2z = 6 \quad \text{(Plane 2)}
$

we follow these steps:

---

### 1. **Find the Direction Vector of the Line**
The direction vector of the line of intersection is the **cross product** of the normal vectors of the two planes.

- The normal vector of Plane 1 is  $ \mathbf{n}_1 = (2, 3, 4) $.
- The normal vector of Plane 2 is $ \mathbf{n}_2 = (3, 4, 2) $.

The cross product $ \mathbf{n}_1 \times \mathbf{n}_2 $ is calculated as:

$
\mathbf{n}_1 \times \mathbf{n}_2 = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 3 & 4 \\
3 & 4 & 2
\end{vmatrix}
$

$
\mathbf{n}_1 \times \mathbf{n}_2 = \mathbf{i}(3 \cdot 2 - 4 \cdot 4) - \mathbf{j}(2 \cdot 2 - 4 \cdot 3) + \mathbf{k}(2 \cdot 4 - 3 \cdot 3)
$

$
\mathbf{n}_1 \times \mathbf{n}_2 = \mathbf{i}(6 - 16) - \mathbf{j}(4 - 12) + \mathbf{k}(8 - 9)
$

$
\mathbf{n}_1 \times \mathbf{n}_2 = -10\mathbf{i} + 8\mathbf{j} - 1\mathbf{k}
$

So, the direction vector of the line is $ \mathbf{d} = (-10, 8, -1) $.

---

### 2. **Find a Point on the Line**
To find a point on the line, solve the system of equations formed by Plane 1 and Plane 2. We can set $ z = 0 $ and solve for \( x \) and \( y \):

$
2x + 3y = 5 \quad \text{(Equation 1)}
$
$
3x + 4y = 6 \quad \text{(Equation 2)}
$

Solve Equation 1 for $ x $:

$
2x = 5 - 3y \implies x = \frac{5 - 3y}{2}
$

Substitute $ x = \frac{5 - 3y}{2} $ into Equation 2:

$
3\left(\frac{5 - 3y}{2}\right) + 4y = 6
$

Multiply through by 2 to eliminate the fraction:

$
3(5 - 3y) + 8y = 12
$

$
15 - 9y + 8y = 12
$

$
15 - y = 12 \implies y = 3
$

Substitute $ y = 3 $ into $ x = \frac{5 - 3y}{2} $:

$
x = \frac{5 - 3(3)}{2} = \frac{5 - 9}{2} = -2
$

So, a point on the line is $ (-2, 3, 0) $.

---

### 3. **Write the Parametric Equations of the Line**
Using the direction vector $ \mathbf{d} = (-10, 8, -1) $ and the point $ (-2, 3, 0) $, the parametric equations of the line are:

The point satisfies both equations, confirming it lies on the line of intersection.

To find examples of planes that are **parallel** and **perpendicular** to the plane $ 2x + 3y + 4z = 5 $, we analyze the normal vector of the given plane.

### 2. **Parallel Plane**
A plane is **parallel** to the given plane if it has the **same normal vector** $ \mathbf{n} = (2, 3, 4) $. The equation of such a plane is:
$
2x + 3y + 4z = k
$
where $ k $ is a constant different from $ 5 $.

**Example:**
Choose $ k = 10 $. The equation of the parallel plane is:
$
2x + 3y + 4z = 10
$

---

### 3. **Perpendicular Plane**
A plane is **perpendicular** to the given plane if its normal vector is **orthogonal** to $ \mathbf{n} = (2, 3, 4) $. This means the dot product of the two normal vectors must be zero.

Let the normal vector of the perpendicular plane be $ \mathbf{m} = (a, b, c) $. Then:
$
\mathbf{n} \cdot \mathbf{m} = 2a + 3b + 4c = 0
$

We can choose any $ a, b, c $ that satisfy this equation. For simplicity, let $ a = 3 $, $ b = -2 $, and $ c = 0 $. Then:
$
2(3) + 3(-2) + 4(0) = 6 - 6 + 0 = 0
$

**Example:**
The equation of the perpendicular plane is:
$
3x - 2y + 0z = k
$
Choose \( k = 1 \). The equation becomes:
$
3x - 2y = 1
$

---

### 4. **Verification**
- **Parallel Plane:** The normal vector of $ 2x + 3y + 4z = 10 $ is $ (2, 3, 4) $, which matches the normal vector of the given plane, confirming that the planes are parallel.
- **Perpendicular Plane:** The normal vector of $ 3x - 2y = 1 $ is $ (3, -2, 0) $. The dot product with \( \mathbf{n} = (2, 3, 4) \) is:

To find the distance from the point $A(1, 2, 3)$ to the plane $2x + 3y + 4z = 5$, we use the **distance formula from a point to a plane**. The formula is:

$$
\text{Distance} = \frac{|ax_0 + by_0 + cz_0 + d|}{\sqrt{a^2 + b^2 + c^2}}
$$

where:
- $(x_0, y_0, z_0)$ are the coordinates of the point $A$,
- $ax + by + cz + d = 0$ is the equation of the plane.

---

### Step 1: Rewrite the Plane Equation
The given plane equation is $2x + 3y + 4z = 5$. Rewrite it in the standard form $ax + by + cz + d = 0$:

$$
2x + 3y + 4z - 5 = 0
$$

Here, $a = 2$, $b = 3$, $c = 4$, and $d = -5$.

---

### Step 2: Substitute into the Distance Formula
The coordinates of point $A$ are $(x_0, y_0, z_0) = (1, 2, 3)$. Substitute these values into the distance formula:

$$
\text{Distance} = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

---

### Step 3: Simplify the Numerator
Calculate the numerator:

$$
2(1) + 3(2) + 4(3) - 5 = 2 + 6 + 12 - 5 = 15
$$

Take the absolute value:

$$
|15| = 15
$$

---

### Step 4: Simplify the Denominator
Calculate the denominator:

$$
\sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29}
$$

---

### Step 5: Compute the Distance
Divide the numerator by the denominator:

$$
\text{Distance} = \frac{15}{\sqrt{29}}
$$

Rationalize the denominator:

$$
\text{Distance} = \frac{15\sqrt{29}}{29}
$$

---

### Final Answer:
The distance from point $A(1, 2, 3)$ to the plane $2x + 3y + 4z = 5$ is:

$$
\boxed{\frac{15\sqrt{29}}{29}}
$$

```python
#9import sympy as sp

# Define the normal vectors
n1 = sp.Matrix([1, 1, 1])  # Normal vector of the plane x + y + z = 1
n2 = sp.Matrix([1, 0, 0])  # Normal vector of the yz-plane (x = 0)

# Calculate the dot product of the normal vectors
dot_product = n1.dot(n2)

# Calculate the magnitudes of the normal vectors
magnitude_n1 = sp.sqrt(sum(component**2 for component in n1))
magnitude_n2 = sp.sqrt(sum(component**2 for component in n2))

# Calculate the cosine of the angle
cos_theta = dot_product / (magnitude_n1 * magnitude_n2)

# Calculate the angle in degrees
theta = sp.acos(cos_theta)
theta_degrees = sp.deg(theta)

# Display the results
print("Cosine of the angle:", cos_theta)
print("Angle in degrees:", theta_degrees)
```
