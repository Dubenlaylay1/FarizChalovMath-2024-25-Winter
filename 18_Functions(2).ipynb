```python
#1
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
    return 3 * x - 1

def g(x):
    return np.sqrt(x)

# Define the compositions of the functions
def fg(x):
    return f(g(x))

def gf(x):
    return g(f(x))

def ff(x):
    return f(f(x))

def gg(x):
    return g(g(x))

# Generate x values for plotting
x_values = np.linspace(0, 10, 400)
x_values_gf = np.linspace(0, 4, 400)  # Adjust range for gf and gg to avoid complex values
x_values_gg = np.linspace(0, 3, 400)

# Plot the functions and their compositions
plt.figure(figsize=(12, 8))

# Plot f(g(x))
plt.subplot(2, 2, 1)
plt.plot(x_values, fg(x_values), label='f(g(x))')
plt.title('f(g(x)) = 3sqrt(x) - 1')
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend()
plt.grid(True)

# Plot g(f(x))
plt.subplot(2, 2, 2)
plt.plot(x_values_gf, gf(x_values_gf), label='g(f(x))')
plt.title('g(f(x)) = sqrt(3x - 1)')
plt.xlabel('x')
plt.ylabel('g(f(x))')
plt.legend()
plt.grid(True)

# Plot f(f(x))
plt.subplot(2, 2, 3)
plt.plot(x_values, ff(x_values), label='f(f(x))')
plt.title('f(f(x)) = 3(3x - 1) - 1')
plt.xlabel('x')
plt.ylabel('f(f(x))')
plt.legend()
plt.grid(True)

# Plot g(g(x))
plt.subplot(2, 2, 4)
plt.plot(x_values_gg, gg(x_values_gg), label='g(g(x))')
plt.title('g(g(x)) = sqrt(sqrt(x))')
plt.xlabel('x')
plt.ylabel('g(g(x))')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return np.exp(x)

def g(x):
    return np.log(x)

# Define the compositions of the functions
def fg(x):
    return f(g(x))

def gf(x):
    return g(f(x))

# Generate x values for plotting
x_values = np.linspace(0.1, 10, 400)  # Avoid zero to prevent log(0)
x_values_inverse = np.linspace(-2, 2, 400)

# Plot the functions and their compositions
plt.figure(figsize=(12, 6))

# Plot f(g(x))
plt.subplot(1, 2, 1)
plt.plot(x_values, fg(x_values), label='f(g(x)) = e^(ln(x)) = x')
plt.plot(x_values, x_values, 'r--', label='y = x')  # Reference line y = x
plt.title('f(g(x)) = x')
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend()
plt.grid(True)

# Plot g(f(x))
plt.subplot(1, 2, 2)
plt.plot(x_values_inverse, gf(x_values_inverse), label='g(f(x)) = ln(e^x) = x')
plt.plot(x_values_inverse, x_values_inverse, 'r--', label='y = x')  # Reference line y = x
plt.title('g(f(x)) = x')
plt.xlabel('x')
plt.ylabel('g(f(x))')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
####Both the capacity f(g(x)) and g(f(x)) are robust to x. This shows that the capacity f(x) and g(x) are functions of their parameters.
```

```python
#4
# Define the given function as a list of tuples
f = [(1, 7), (2, 7), (3, 11)]

# Check if the function is one-to-one
def is_one_to_one(function):
    outputs = [output for _, output in function]
    # A function is one-to-one if all outputs are unique
    return len(outputs) == len(set(outputs))

# Create the "inverse" of the function by switching input and output
def create_inverse(function):
    return [(output, input) for input, output in function]

# Check if the inverse is a valid function
def is_valid_function(inverse):
    inputs = [input for input, _ in inverse]
    # A valid function must have unique inputs
    return len(inputs) == len(set(inputs))

# Check if the given function is one-to-one
if is_one_to_one(f):
    print("The function is one-to-one and has an inverse.")
    inverse_f = create_inverse(f)
    print("The inverse function is:", inverse_f)
else:
    print("The function is not one-to-one and does not have an inverse.")

# Demonstrate the invalid inverse in this case
inverse_f = create_inverse(f)
if not is_valid_function(inverse_f):
    print("If we try to switch inputs and outputs, we get:")
    print("Inverse attempt:", inverse_f)
    print("This is not a valid function because one input maps to multiple outputs.")
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) and its inverse f^(-1)(x)
def f(x):
    return x - 1

def f_inverse(x):
    return x + 1

# Generate x values for plotting
x = np.linspace(-10, 10, 500)

# Evaluate both functions
y_f = f(x)
y_f_inverse = f_inverse(x)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot f(x)
plt.plot(x, y_f, label=r"$f(x) = x - 1$", color="blue")

# Plot f^(-1)(x)
plt.plot(x, y_f_inverse, label=r"$f^{-1}(x) = x + 1$", color="orange")

# Plot the identity line y = x for reference
plt.plot(x, x, label="y = x (Identity Line)", color="green", linestyle="--")

# Add labels, legend, and title
plt.title("Function and Its Inverse")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
