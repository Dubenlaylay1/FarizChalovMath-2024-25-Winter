### 10.

We start with the integral:
$
\int \sqrt{10x} \, dx
$

#### Step 1: Factor out the constant
$
\int \sqrt{10x} \, dx = \sqrt{10} \int \sqrt{x} \, dx
$

#### Step 2: Express the square root as a power
$
\sqrt{10} \int \sqrt{x} \, dx = \sqrt{10} \int x^{1/2} \, dx
$

#### Step 3: Integrate the power expression
$
\sqrt{10} \int x^{1/2} \, dx = \sqrt{10} \left[\frac{x^{1/2 + 1}}{1/2 + 1}\right] + C
$

#### Step 4: Simplify
$
\sqrt{10} \left[\frac{x^{3/2}}{3/2}\right] + C = \frac{2}{3} \sqrt{10} \, x^{3/2} + C
$

### Final Result:
$
\int \sqrt{10x} \, dx = \frac{2}{3} \sqrt{10} \, x^{3/2} + C
$

### Solution

We start with the integral:
$
\int \cos\left(\frac{5}{2}x + 3\right) dx
$

#### Step 1: Substitution
Let:
$
u = \frac{5}{2}x + 3
$
Then:
$
\frac{du}{dx} = \frac{5}{2} \quad \text{or} \quad dx = \frac{2}{5} du
$

#### Step 2: Rewrite the integral with the new variable
$
\int \cos\left(\frac{5}{2}x + 3\right) dx = \int \cos(u) \cdot \frac{2}{5} \, du
$

#### Step 3: Factor out the constant
$
\frac{2}{5} \int \cos(u) \, du
$

#### Step 4: Integrate \(\cos(u)\)
The integral of \(\cos(u)\) is \(\sin(u)\):
$
\frac{2}{5} \sin(u) + C
$

#### Step 5: Substitute \(u\) back
$
\frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
$

### Final Result:
$
\int \cos\left(\frac{5}{2}x + 3\right) dx = \frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
$

$
\sin(u) + C

$

### Solution

We start with the integral:
$
\int x \ln(x) \, dx
$

#### Step 1: Identify \( u \) and \( dv \) for integration by parts
Let:

$ u = \ln(x) \quad \text{and} \quad dv = x \, dx
$

Then:
$
du = \frac{1}{x} \, dx \quad \text{and} \quad v = \frac{x^2}{2}
$

#### Step 2: Use the integration by parts formula
The formula is:
$
\int u \, dv = uv - \int v \, du
$

Substitute $ u = \ln(x) $, $ du = \frac{1}{x} dx $, $ v = \frac{x^2}{2} $, and $ dv = x \, dx $:
$
\int x \ln(x) \, dx = \ln(x) \cdot \frac{x^2}{2} - \int \frac{x^2}{2} \cdot \frac{1}{x} \, dx
$

Simplify the second term:
$
\int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \int \frac{x^2}{2x} \, dx
$

$
= \frac{x^2}{2} \ln(x) - \int \frac{x}{2} \, dx
$

#### Step 3: Integrate the remaining term
$
\int \frac{x}{2} \$, dx = \frac{1}{2} \int x \, dx = \frac{1}{2} \cdot \frac{x^2}{2} = \frac{x^2}{4}
$

So:
$
\int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \frac{x^2}{4}
$

#### Step 4: Simplify and add the constant of integration
Factor out $ \frac{x^2}{4} $:
$
\int x \ln(x) \, dx = \frac{x^2}{4} \left(2 \ln(x) - 1\right) + C
$

---

### Final Result:
$
\int x \ln(x) \, dx = \frac{x^2}{4} \left(2 \ln(x) - 1\right) + C
$

$
u = x \quad \text{and} \quad dv = e^x \, dx
$

Then:
$
du = dx \quad \text{and} \quad v = e^x
$

#### Step 2: Use the integration by parts formula
The formula is:
$
\int u \, dv = uv - \int v \, du
$

Substitute \( u = x \), \( du = dx \), \( v = e^x \), and \( dv = e^x \, dx \):
$
\int x e^x \, dx = x e^x - \int e^x \, dx
$

#### Step 3: Integrate \( e^x \)
The integral of \( e^x \) is itself:
$
\int e^x \, dx = e^x
$

So:
$
\int x e^x \, dx = x e^x - e^x + C
$

#### Step 4: Factor out \( e^x \)
$
\int x e^x \, dx = e^x (x - 1) + C
$

---

### Final Result:
$
\int x e^x \, dx = e^x (x - 1) + C

$
\pi^2 + \pi

0^2 + 0 = 0

### Solution

We need to calculate the area of the region bounded by the lines x = 1 ,  x = 2 ,  y = 0 , and $ y = x^2 + 1 $. This is done by setting up a definite integral. The function describing the upper boundary of the region is $ y = x^2 + 1 $, and the lower boundary is  y = 0.

#### Step 1: Set up the integral
The area is given by the integral of the difference between the upper curve and the lower curve over the interval  [1, 2] :
$
\text{Area} = \int_{1}^{2} \left( (x^2 + 1) - 0 \right) dx
$
$
\text{Area} = \int_{1}^{2} (x^2 + 1) \, dx
$

#### Step 2: Integrate the function
Now, we compute the integral:
$
\int (x^2 + 1) \, dx = \int x^2 \, dx + \int 1 \, dx = \frac{x^3}{3} + x
$

#### Step 3: Evaluate the definite integral
We now apply the limits of integration \( x = 1 \) and \( x = 2 \):

$
\left[ \frac{x^3}{3} + x \right]_{1}^{2}
$

Evaluate at \( x = 2 \):
$
\frac{2^3}{3} + 2 = \frac{8}{3} + 2 = \frac{8}{3} + \frac{6}{3} = \frac{14}{3}
$

Evaluate at \( x = 1 \):
$
\frac{1^3}{3} + 1 = \frac{1}{3} + 1 = \frac{1}{3} + \frac{3}{3} = \frac{4}{3}
$

#### Step 4: Subtract the results
Now subtract the value at \( x = 1 \) from the value at \( x = 2 \):

$
\text{Area} = \frac{14}{3} - \frac{4}{3} = \frac{10}{3}
$

---

### Final Result:
The area of the region bounded by the lines \( x = 1 \), \( x = 2 \), \( y = 0 \), and \( y = x^2 + 1 \) is:
$
\frac{10}{3} \text{ square units}
$

$
\int \sin(x) \, dx = -\cos(x)
$

Now, we evaluate the definite integral over the interval $ [0, \pi] $:

$
P = \left[ -\cos(x) \right]_{0}^{\pi}
$

#### Step 3: Evaluate at the upper and lower limits
First, evaluate at $x = \pi $:
$
-\cos(\pi) = -(-1) = 1
$

$
-\cos(0) = -(1) = -1
$

#### Step 4: Subtract the results
Now, subtract the value at \( x = 0 \) from the value at $ x = \pi $:
$
P = 1 - (-1) = 1 + 1 = 2
$

$
L = \int_{a}^{b} \sqrt{1 + \left(f'(x)\right)^2} \, dx
$

f'(x) = \cos(x)

L = \int_{0}^{\pi} \sqrt{1 + \left( \cos(x) \right)^2} \, dx

### 6

We are given the position function of a particle:

$
x(t) = 3t^2 - 6t + 1
$

We are asked to find the distance traveled by the particle between \( t = 0 \) and \( t = 2 \).

#### Step 1: Find the velocity function
The velocity is the derivative of the position function  x(t) :

$
v(t) = \frac{dx}{dt} = \frac{d}{dt} \left( 3t^2 - 6t + 1 \right)
$

Taking the derivative:

$
v(t) = 6t - 6
$

#### Step 2: Find when the velocity changes sign
We solve  v(t) = 0  to find when the particle changes direction:

$
6t - 6 = 0 \quad \Rightarrow \quad t = 1
$

Thus, the particle changes direction at \( t = 1 \).

#### Step 3: Calculate the distance traveled in two intervals
We will calculate the distance traveled in the two intervals:
- From \( t = 0 \) to \( t = 1 \)
- From \( t = 1 \) to \( t = 2 \)

#### Step 4: Integrate the absolute value of the velocity
The distance traveled is the integral of the speed, which is the absolute value of the velocity \( |v(t)| \). We can split the integral at \( t = 1 \) and compute each part:

$
\text{Distance} = \int_{0}^{1} |v(t)| \, dt + \int_{1}^{2} |v(t)| \, dt
$

- For $ 0 \leq t \leq 1 $, $ v(t) = 6t - 6 \leq 0 $, so $ |v(t)| = -(6t - 6) = 6 - 6t $.
- For $ 1 \leq t \leq 2 $, $ v(t) = 6t - 6 \geq 0 $, so $ |v(t)| = 6t - 6 $.

#### Calculate each integral:

##### From \( t = 0 \) to \( t = 1 \):

$
\int_{0}^{1} (6 - 6t) \, dt = \left[ 6t - 3t^2 \right]_{0}^{1} = (6 - 3) - (0) = 3
$

##### From \( t = 1 \) to \( t = 2 \):

$
\int_{1}^{2} (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_{1}^{2} = \left( 3(2^2) - 6(2) \right) - \left( 3(1^2) - 6(1) \right)
$

$
= (12 - 12) - (3 - 6) = 0 - (-3) = 3
$

#### Total Distance:
The total distance traveled is the sum of the two integrals:

$
\text{Distance} = 3 + 3 = 6
$

---

### Final Result:
The distance traveled by the particle between  t = 0  and  t = 2  is:

$$
\boxed{6}
$$
